## 3. Evaluation

```python
# Placeholder – replace with real predictions
y_true = np.array([0, 0, 1, 1, 1, 0])
y_pred = np.array([0, 0, 1, 0, 1, 0])

print(classification_report(y_true, y_pred, target_names=['Genuine', 'Impostor']))
far = compute_far(y_true, y_pred)
print(f'False Acceptance Rate (FAR): {far:.4f}')
```

## 4. Confusion Matrix

```python
import seaborn as sns

cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Genuine', 'Impostor'],
            yticklabels=['Genuine', 'Impostor'])
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()
```
